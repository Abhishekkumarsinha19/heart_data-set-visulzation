```python
#Line Plot
#Visualize trends or changes
import matplotlib.pyplot as plt

plt.plot(df['age'], df['thalach'], label='Max Heart Rate vs Age', color='blue', marker='o')
plt.title('Max Heart Rate (thalach) vs Age')
plt.xlabel('Age')
plt.ylabel('Max Heart Rate (thalach)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#Scatter Plot
#Visualize relationships between two columns:
plt.scatter(df['age'], df['chol'], color='red', alpha=0.5)
plt.title('Cholesterol Levels vs Age')
plt.xlabel('Age')
plt.ylabel('Cholesterol (chol)')
plt.grid(True)
plt.show()
```

```python
#Histogram
plt.hist(df['age'], bins=10, color='purple', alpha=0.7)
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
#Bar Plot
cp_counts = df['cp'].value_counts()
plt.bar(cp_counts.index, cp_counts.values, color='orange')
plt.title('Chest Pain Type Distribution')
plt.xlabel('Chest Pain Type (cp)')
plt.ylabel('Count')
plt.xticks(cp_counts.index)
plt.show()
```

```python
#Pie Chart
target_counts = df['target'].value_counts()
plt.pie(target_counts, labels=['No Disease', 'Disease'], autopct='%1.1f%%', startangle=90, colors=['green', 'red'])
plt.title('Heart Disease Distribution')
plt.show()
```

```python
#Box Plot
plt.boxplot(df['trestbps'])
plt.title('Resting Blood Pressure Distribution')
plt.ylabel('Resting Blood Pressure (trestbps)')
plt.show()
```

```python
sns.heatmap(df.isnull(), cbar=False, cmap='viridis')
plt.title('Missing Data Heatmap')
plt.show()
```

```python
#Check correlations between numerical features
plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Age Distribution
sns.histplot(df['age'], kde=True, bins=20, color='blue')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
#Cholesterol Distribution by Target
sns.kdeplot(data=df, x='chol', hue='target', fill=True, palette='muted')
plt.title('Cholesterol Distribution by Target')
plt.show()
```

```python
#Count of Chest Pain Type
sns.countplot(data=df, x='cp', hue='target', palette='Set2')
plt.title('Chest Pain Type Count by Target')
plt.show()
```

```python
#Bar Plot: Max Heart Rate by Sex
sns.barplot(data=df, x='sex', y='thalach', ci='sd', palette='muted')
plt.title('Average Max Heart Rate by Sex')
plt.xlabel('Sex (0 = Female, 1 = Male)')
plt.ylabel('Max Heart Rate (thalach)')
plt.show()
```

```python
#Box Plot: Resting Blood Pressure
sns.boxplot(data=df, y='trestbps', color='orange')
plt.title('Resting Blood Pressure Distribution')
plt.ylabel('Resting Blood Pressure (trestbps)')
plt.show()
```

```python
#Violin Plot: Max Heart Rate by Target
sns.violinplot(data=df, x='target', y='thalach', palette='pastel')
plt.title('Max Heart Rate Distribution by Target')
plt.xlabel('Target (0 = No Disease, 1 = Disease)')
plt.ylabel('Max Heart Rate (thalach)')
plt.show()
```

```python
#Scatter Plot: Max Heart Rate vs Age
sns.scatterplot(data=df, x='age', y='thalach', hue='target', style='sex', palette='coolwarm')
plt.title('Max Heart Rate vs Age by Target and Sex')
plt.show()
```

```python
#Joint Plot: Age vs Cholesterol
sns.jointplot(data=df, x='age', y='chol', kind='scatter', hue='target', palette='Set1')
plt.show()
```

```python
#Pair Plot: Selected Features
sns.pairplot(df[['age', 'chol', 'thalach', 'target']], hue='target', palette='Set2')
plt.show()
```

```python
#Strip Plot: ST Depression (oldpeak) by Slope
sns.stripplot(data=df, x='slope', y='oldpeak', hue='target', jitter=True, palette='Set2')
plt.title('ST Depression by Slope and Target')
plt.show()
```

```python
#Swarm Plot: Cholesterol by Chest Pain Type
sns.swarmplot(data=df, x='cp', y='chol', hue='sex', palette='deep')
plt.title('Cholesterol Levels by Chest Pain Type and Sex')
plt.show()
```

```python
# FacetGrid: Age vs Max Heart Rate by Sex
g = sns.FacetGrid(df, col='sex', hue='target', height=4, aspect=1.2)
g.map(sns.scatterplot, 'age', 'thalach')
g.add_legend()
plt.show()
```

```python
#Regression Plot: Age vs Max Heart Rate
sns.regplot(data=df, x='age', y='thalach', scatter_kws={'alpha':0.5}, color='green')
plt.title('Linear Relationship between Age and Max Heart Rate')
plt.show()
```

```python
#Violin Plot by Chest Pain Type
sns.catplot(data=df, x='cp', y='thalach', hue='target', kind='violin', split=True, height=5, aspect=1.5)
plt.title('Max Heart Rate by Chest Pain Type and Target')
plt.show()
```

```python
#Relational Plot: Max Heart Rate vs Age
sns.relplot(data=df, x='age', y='thalach', hue='sex', style='target', kind='scatter', size='chol', height=6)
plt.title('Max Heart Rate vs Age by Sex and Target')
plt.show()
```
